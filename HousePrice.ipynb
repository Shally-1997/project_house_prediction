```python
df1['sale_date'] = pd.to_datetime(df1['sale_date'], errors='coerce')
print(df1['sale_date'].dtype, df1['sale_date'].isnull().sum(), "null dates")
```

```python
null_counts = df1.isnull().sum()
null_pct    = df1.isnull().mean() * 100

summary_nulls = pd.DataFrame({
    'null_count': null_counts,
    'null_pct':   null_pct.round(2)
}).sort_values('null_pct', ascending=False)
print(summary_nulls)
```

```python
numeric = df1.select_dtypes(include='number')
ranges  = numeric.agg(['min','max']).T.rename(columns={'min':'Min','max':'Max'})
print(ranges)
```
